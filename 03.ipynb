```python
import math
import numpy as np

def switcher(element,n):#nに線として表示したい数字を入れる
    global grid #関数の外の変数を関数内で変える
    global posX
    global posY
    if 'U' in element:#引数のなかに'U'があるか
        limit = int(element[1:])#先頭以外の文字を抽出しつつintで数値に変換
        for i in range (posY,posY+limit+1):
            if grid[i][posX] == n or grid[i][posX] == 0:
                grid[i][posX] = n
            else:
                grid[i][posX] += n
        posY += limit#現在地の更新
    elif 'D' in element:
        limit = int(element[1:])
        for i in range (posY-limit,posY+1):
            if grid[i][posX] == n or grid[i][posX] == 0:
                grid[i][posX] = n 
            else:
                grid[i][posX] += n
        posY -= limit
    elif 'R' in element:
        limit = int(element[1:])
        for i in range (posX,posX+limit +1):
            if grid[posY][i] == n or grid[posY][i] == 0:
                grid[posY][i] = n
            else:
                grid[posY][i] += n
        posX += limit
    elif 'L' in element:
        limit = int(element[1:])
        for i in range (posX-limit,posX+1):
            if grid[posY][i] == n or grid[posY][i] == 0:
                grid[posY][i] = n
            else:
                grid[posY][i] += n
        posX -= limit

def distance():#最短距離を図る
    pos =np.zeros((2*yoko,2))
    son = np.where(grid ==3)#重なっているところのインデックスを返す
    a = son[0]#縦座標
    b = son[1]#横座標
    distance = np.zeros((len(a),2))
    oke =np.zeros(len(a))
    for i in range (len(a)):#座標を(x,y)の形で入れる
        distance[i] = (a[i],b[i])   
    for i in range (len(distance)):#原点からの距離を絶対値で配列に入れる
        oke[i] = abs(distance[i][0] - yoko + 1) + abs(distance[i][1] - tate + 1)
    ma = np.argmin(oke) #配列には原点の座標も含まれている　最小値のインデックスを返す　
    oke[ma] = np.amax(oke)#最小値を最大値に置き換える
    print(np.amin(oke),'dayo')#原点の次に距離が短いやつの距離

f = open('input.txt','r')
a = f.readline()
b = f.readline()
a = a.split(',')
b = b.split(',')


tate = 20000
yoko = 20000

  
#grid =[[0 for i in range(2*yoko)] for j in range(2*tate)]
grid =np.zeros((2*yoko,2*tate))
posX = yoko-1 #current X position
posY = tate-1 #current Y position  
print('koko')
for j in a:
    switcher(j,2)  

posX = yoko-1 #current X position
posY = tate-1 #current Y position
print('koko1')
for j in b:
    switcher(j,1)

gyaku = [0 for i in range(2*tate)]
for i in range (2*tate):
    gyaku[i] = 2*tate-i-1

distance()
```
